```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# 문장을 임베딩으로 변환
model_name = 'sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2'
model = SentenceTransformer(model_name)
```

```python
embedding1 = model.encode(df_100['제목_spacing'].to_list())
embedding2 = model.encode(task['단위과제설명'].to_list())
```

```python
print(embedding1)
print('=========')
print(embedding2)
```

```python
'''
embedding1[0] = 차량배차신청서
embedding1[1] = 차량운행일지
embedding1[2] = [시설환경팀] 12월 추가 교육시간 반영 요청
embedding1[3] = [시설환경팀]2024년도 삼다수사업장 공병부산물 매각 온비드 입찰 협조요청
'''

# 첫 번째와 두 번째 간의 유사도
similarity_1_2 = cosine_similarity([embedding1[0]], [embedding1[1]])
print("첫 번째와 두 번째 간의 유사도:", similarity_1_2)

# 첫 번째와 세 번째 간의 유사도
similarity_1_3 = cosine_similarity([embedding1[0]], [embedding1[2]])
print("첫 번째와 세 번째 간의 유사도:", similarity_1_3)

# 두 번째와 세 번째 간의 유사도
similarity_2_3 = cosine_similarity([embedding1[1]], [embedding1[2]])
print("두 번째와 세 번째 간의 유사도:", similarity_2_3)

# 세 번째와 네 번째 간의 유사도
similarity_3_4 = cosine_similarity([embedding1[2]], [embedding1[3]])
print("두 번째와 세 번째 간의 유사도:", similarity_3_4)
```

### 부서 열 생성

```python
# '부서' 열 생성
df_100['부서'] = df_100['문서번호'].apply(lambda x: x.split('-')[0])

print(df_100['부서'].tail(10))
```

```python
rematching_task = []

# embedding1의 모든 행 반복
for i, first_embedding in enumerate(embedding1):
    # 해당 인덱스의 '부서' 추출
    department = df_100.loc[i, '부서']

    # 최대 유사도와 해당 인덱스 저장 변수
    max_similarity = -1
    max_index = -1

    # task의 모든 행 반복
    task_cnt = 0
    for j, task_row in task.iterrows():
        # '부서명'이 '부서공통'의 값과 일치하는 경우에만 유사도 계산
        if task_row['부서명'] == department or task_row['부서명'] == '부서공통':
            # 코사인 유사도 계산
            similarity = np.dot(first_embedding, embedding2[j]) / (np.linalg.norm(first_embedding) * np.linalg.norm(embedding2[j]))
            task_cnt += 1
            # 최대 유사도 갱신
            if similarity > max_similarity:
                max_similarity = similarity
                max_index = j

    # 결과 출력
    print(f"[첫번째 행 {i+1}] 최대 유사도: {max_similarity:.4f}, 해당 인덱스: {max_index}, task의 부서명 개수: {task_cnt}, 부서별 task 개수: {len(task[task['부서명'] == department])+len(task[task['부서명'] == '부서공통'])}")


    # rematching_task 리스트에 최대 유사도를 갖는 task의 인덱스 추가
    rematching_task.append(max_index)

# rematching_task를 사용하여 df_100의 '부서'와 task의 '부서명'이 '부서공통'의 값과 일치하는 단위과제명을 할당
df_100['할당된_단위과제명'] = task.loc[rematching_task, '단위과제명'].values
```

```python
df_100['할당된_단위과제명'].head(100)
```

```python
# 정확하게 일치하는 횟수 세기
correct_count = sum(df_100['단위업무명'] == df_100['할당된_단위과제명'])

# 전체 예측의 수
total_predictions = len(df_100)

# 정확도 계산
accuracy = correct_count / total_predictions

print("정확도:", accuracy)
print(f'맞춘 개수 / 전체 개수 : {correct_count}/{total_predictions}')
```
